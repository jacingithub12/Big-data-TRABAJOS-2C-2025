<H3> 2. Manejo de datos con Pandas 久엪n
Pandas es una librer칤a fundamental para trabajar con datos tabulares. Un DataFrame de Pandas es una estructura similar a una hoja de c치lculo, con filas y columnas.

Crear un DataFrame

Podemos crear un DataFrame a partir de listas de Python.

```python
# Listas de datos
lista_personajes = ['Luke Skywalker', 'C-3PO', 'Darth Vader']
lista_altura = [172, 167, 202]

# Crear el DataFrame
import pandas as pd

d_star_wars = pd.DataFrame({
    'personaje': lista_personajes,
    'altura': lista_altura
})

# Mostrar el DataFrame y sus tipos de datos
print(d_star_wars)
print("---")
print(d_star_wars.dtypes)
```

<H3>4. Filtrar y seleccionar datos 久쓮n
Para seleccionar datos espec칤ficos de tu DataFrame, usa loc. Con loc puedes seleccionar filas y columnas por sus nombres o etiquetas.


El formato loc[filas, columnas] te permite seleccionar datos usando las etiquetas de filas y columnas.

[0, :]: El primer valor (0) selecciona la fila con la etiqueta 0. Los dos puntos (:) significan todas las columnas.

[:, 'nombre_columna']: Los dos puntos (:) seleccionan todas las filas. El nombre de la columna selecciona una columna espec칤fica.

4.1. Seleccionar columnas

```python
# Seleccionar una sola columna
#d_star_wars['personaje']

# Seleccionar m칰ltiples columnas
d_star_wars[['personaje', 'altura']]
```

4.2. Seleccionar filas

Podemos seleccionar filas bas치ndote en su posici칩n (칤ndice).

Nota: el uso de 0:2 en loc es inclusivo, lo que significa que incluye los 칤ndices 0, 1 y 2.

```python
# Seleccionar la primera fila
d_star_wars.loc[0, :]
```

```python
# Seleccionar las 3 primeras filas
d_star_wars.loc[0:2, :]
```

4.3. Filtrar con condiciones

Podemos usar una condici칩n para filtrar filas que cumplan con un criterio espec칤fico. Por ejemplo, encontrar todos los personajes cuya altura sea mayor que 200.

```python
# Crear un filtro (devuelve True/False para cada fila)
filtro_altura = d_star_wars['altura'] > 200

# Aplicar el filtro para obtener solo los registros que cumplen la condici칩n
d_star_wars.loc[filtro_altura, :]
```

4.4. Reemplazar valores

Podemos combinar el filtrado con la asignaci칩n para cambiar valores espec칤ficos en el DataFrame.

```python
# Reemplazar la altura de Darth Vader por 185
d_star_wars.loc[d_star_wars['personaje'] == 'Darth Vader', 'altura'] = 185

# Verificar el cambio
print(d_star_wars.loc[d_star_wars['personaje'] == 'Darth Vader', :])
```

<H3>5. Manejar valores nulos (NaN) 久꾚n
5.1. Los valores faltantes (NaN) son muy comunes. Es crucial saber c칩mo identificarlos y manejarlos.

```python
# Identificar nulos por columna
print("N칰mero de nulos en cada columna:")
print(d_star_wars_url.isnull().sum())
```

```python
d_star_wars_url
```

```python
# Reemplazar valores nulos en columnas categ칩ricas
valores_a_rellenar = {'homeworld': 'nulo_world', 'species': 'nulo_species'}

d_star_wars_url.fillna(value=valores_a_rellenar, inplace=True)
print(d_star_wars_url)
```

```python
#Vemos cual es el promedio de altura
print(d_star_wars_url['height'].mean())
```

```python
# Reemplazar valores nulos en columnas num칠ricas con el promedio
promedio_altura = d_star_wars_url['height'].mean()


d_star_wars_url['height'] = d_star_wars_url['height'].fillna(promedio_altura)
d_star_wars_url
```

```python
# Verificar que los nulos fueron reemplazados
status(d_star_wars_url['height'])
```

5.2. Eliminar registros con valores nulos
A veces, en lugar de reemplazar los valores nulos, lo mejor es eliminar por completo las filas o columnas que los contienen. El m칠todo dropna() de Pandas te permite hacerlo f치cilmente.

Por defecto, dropna() elimina cualquier fila que contenga al menos un valor nulo. Es una forma r치pida de limpiar el conjunto de datos si las filas con valores faltantes no son importantes para tu an치lisis.

```python
# Eliminar las filas con valores nulos
d_star_wars_url.dropna(inplace=True)
status(d_star_wars_url)
```

<H3> 6. Crear nuevas columnas 游띎

Puedes crear una nueva columna en un DataFrame y asignarle valores basados en una condici칩n. Es una forma muy poderosa de derivar informaci칩n a partir de los datos existentes.

```python
# Asignamos True si la altura es > 180, de lo contrario, False
d_star_wars['is_tall'] = d_star_wars['altura'] > 180

# Muestra el DataFrame con la nueva columna
print(d_star_wars)
```

<H3> 7. Copiar un DataFrame 춸

Cuando realizas cambios en un DataFrame y quieres conservar la versi칩n original, es una buena pr치ctica hacer una copia. Esto te permite manipular la nueva versi칩n sin afectar la original.

```python
#Hacer una copia del DataFrame
d_star_wars_copy = d_star_wars.copy()

# Ahora puedes trabajar con d_star_wars_copy sin afectar el original
d_star_wars_copy
```
